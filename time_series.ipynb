```python
import os
import pandas as pd

# Folder containing the CSV files
folder_path = 'data'

# List to hold individual DataFrames
dfs = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df.head()
```

```python
# Extract start datetime string (before " - ")
start_times = combined_df['MTU (CET/CEST)'].str.split(' - ').str[0]

# Remove any " (CET)" or " (CEST)" suffix if present
start_times_clean = start_times.str.replace(r' \(CET\)| \(CEST\)', '', regex=True)

# Now parse cleaned datetime strings with dayfirst=True (since format is dd/mm/yyyy)
df['Start DateTime'] = pd.to_datetime(start_times_clean, dayfirst=True)

# Define filter range
start_date = pd.to_datetime('2021-05-21')
end_date = pd.to_datetime('2025-04-30')

# Filter rows within the range inclusive
filtered_df = df[(df['Start DateTime'] >= start_date) & (df['Start DateTime'] <= end_date)]

# Drop helper column if you want
filtered_df = filtered_df.drop(columns=['Start DateTime'])

filtered_df.head()

```

```python
filtered_df.to_parquet('filtered_data.parquet', index=False)
```
